```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
draft = pd.read_csv('draftPER.csv')
draft.head()
```

```python
def rating(x):
    return x[7]+x[8]+x[9]

draft['RTG'] = draft.apply(rating, axis=1)
draft.head()
```

```python
avg = draft.groupby('Pk').mean()
avg = avg.reset_index(drop=False)
avg.index+=1
avg.head()
```

```python
count = draft['Pk'].value_counts()
avg['count']= count
avg.head()
```
