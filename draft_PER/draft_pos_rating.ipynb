```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
```

```python
draft = pd.read_csv('draftPER.csv')
draft.head()
```

```python
def rating(x):
    return x[7]+x[8]+x[9]

draft['RTG'] = draft.apply(rating, axis=1)
draft.head()
```

```python
avg = draft.groupby('Pk').mean()
avg = avg.reset_index(drop=False)
avg.index+=1
avg.head()
```

```python
avg['count']= draft['Pk'].value_counts()
avg.head()
```

```python
avg['Star'] = draft[draft.RTG >=20].groupby('Pk').RTG.count()
avg['Solid'] = draft[(draft.RTG >=15) & (draft.RTG <20)].groupby('Pk').RTG.count()
avg['RoleP'] = draft[(draft.RTG >=10) & (draft.RTG <15)].groupby('Pk').RTG.count()
avg['BenchP'] = draft[(draft.RTG >=5) & (draft.RTG <10)].groupby('Pk').RTG.count()
avg['Bust'] = draft[(draft.RTG >0) & (draft.RTG <5)].groupby('Pk').RTG.count()
avg['DNP'] = draft[pd.isnull(draft.RTG)].groupby('Pk').Pk.count()
avg.head()
```

```python
avg['Star'] = avg['Star']/avg['count']
avg['Solid'] = avg['Solid']/avg['count']
avg['RoleP'] = avg['RoleP']/avg['count']
avg['BenchP'] = avg['BenchP']/avg['count']
avg['Bust'] = avg['Bust']/avg['count']
avg['DNP'] = avg['DNP']/avg['count']
avg = np.round(avg, decimals=3)
avg
#avg.to_csv('Expected_Performance_By_Pick.csv')
```

```python
avg.to_csv('Expected_Performance_By_Pick.csv')
```

```python
pk_rtg = avg[['Pk', 'RTG']]
draft = pd.merge(draft, pk_rtg, on='Pk', how='outer')
draft.head()
```

```python
draft.rename(columns={'RTG_x':'RTG', 'RTG_y':'RTG_proj'}, inplace=True)
draft.head()
```

```python
def rtgdiff(x):
    return x['RTG']-x['RTG_proj']

draft['RTG_diff'] = draft.apply(rtgdiff, axis=1)
draft.head()
```

```python
best_drafting_teams = draft[['Tm','RTG_diff']].groupby('Tm').agg(['mean', 'count'])
best_drafting_teams.columns=best_drafting_teams.columns.droplevel()
best_drafting_teams.reset_index(inplace=True)
best_drafting_teams.rename(columns={'mean':'Team Draft RTG', 'count':'picks'}, inplace=True)
best_drafting_teams.sort(['Team Draft RTG'],ascending=False)
best_drafting_teams.set_index(keys='Tm',inplace=True)
best_drafting_teams
```

```python
best_drafting_teams['Star'] = draft[draft.RTG >=20].groupby('Tm').RTG.count()
best_drafting_teams['Solid'] = draft[(draft.RTG >=15) & (draft.RTG <20)].groupby('Tm').RTG.count()
best_drafting_teams['RoleP'] = draft[(draft.RTG >=10) & (draft.RTG <15)].groupby('Tm').RTG.count()
best_drafting_teams['BenchP'] = draft[(draft.RTG >=5) & (draft.RTG <10)].groupby('Tm').RTG.count()
best_drafting_teams['Bust'] = draft[(draft.RTG >0) & (draft.RTG <5)].groupby('Tm').RTG.count()
best_drafting_teams['DNP'] = draft[pd.isnull(draft.RTG)].groupby('Tm').Tm.count()
```

```python
best_drafting_teams.reset_index(inplace=True)
best_drafting_teams.sort(['Team Draft RTG'],ascending=False,inplace=True)
best_drafting_teams['Team Draft RTG'] = np.round(best_drafting_teams['Team Draft RTG'], decimals=3)
best_drafting_teams
```

```python
best_drafting_teams.to_csv('Best_Drafting_Teams.csv')
```
